```
# Находит имя, на который записан документ
def document_name(number):
    # Перебирает каждый документ
    for doc in documents:
        # Если номер документа совпадает с набранным пользователем номером
        if doc['number'] == number:
            return doc['name']  # Возвращает имя владельца документа
    return 'Документ не найден'  # Возвращает если не найден владелец


# Находит полку, на которой хранится документ
def polka(number):
    # Перебирает каждую полку и документ/документы на полке
    for i, dir in directories.items():
        # Если на полке найден документ
        if number in dir:
            return f"Документ хранится на полке: {i}"  # Возвращает номер полки, на которой находится документ
    return "Документ не найден"  # Если на полках не найден документ


# Все документы
documents = \
    [
        {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
        {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
        {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
    ]
# Полки с документами
directories = \
    {
        '1': ['2207 876234', '11-2'],
        '2': ['10006'],
        '3': []
    }

# Начало цикла
while True:
    # Перечисление всех названий команд
    print("p - имя документа \n"
          "s - полка, в которой лежит документ \n"
          "q - выход")
    # Ввод команды
    command = input("Введите команду: \n")
    # Если пользователь вписал q
    if command == 'q':
        break  # Выход из цикла
    # Если пользователь вписал p
    elif command == 'p':
        number = input("Введите номер:")  # Ввод номера документа
        print(document_name(number))  # Выводит имя владельца документа
    # Если пользователь вписал s
    elif command == 's':
        number = input("Введите номер:")  # Ввод номера документа
        print(polka(number))  # Выводит номер полки, где лежит документ
    # Если пользователь вписал что - то другое
    else:
        print("Такой команды нет. Введите команду заново")  # Выводит, что нет такой команды
```
