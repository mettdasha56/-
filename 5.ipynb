```
from datetime import datetime #импорт datime

#Переводит даты в формате день недели - день - месяц - год
d1 =datetime.strptime('Wednesday, October 2, 2002', '%A, %B %d, %Y')
d2 = datetime.strptime('Friday, 11.10.13', '%A, %d.%m.%y')
d3 = datetime.strptime('Thursday, 18 August 1977', '%A, %d %B %Y')

#Выводит название газеты: день недели, день - месяц - год
print(f"The Moscow Times: {d1.strftime('%A, %d-%m-%Y')}\n"
      f"The Guardian: {d2.strftime('%A, %d-%m-%Y')}\n"
      f"Daily News: {d3.strftime('%A, %d-%m-%Y')}")
```
