```
import json
import csv

categories = {}
with open('purchase_log.txt', 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        data = json.loads(line)
        user_id = data['user_id']
        category = data['category']
        categories[user_id] = category

with open('visit_log__1.csv', 'r', encoding='utf-8') as visits, \
        open('funnel.csv', 'w', encoding='utf-8', newline='') as funnel:
    visits_reader = csv.reader(visits)
    funnel_writer = csv.writer(funnel)

    funnel_writer.writerow(['user_id', 'source', 'category'])

    for row in visits_reader:
        if len(row) < 2:
            continue

        user_id = row[0]
        source = row[1]

        if user_id in categories:
            funnel_writer.writerow([user_id, source, categories[user_id]])
```
